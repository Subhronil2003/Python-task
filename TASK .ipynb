What are identifiers, list rules of identifiers in python ?

ans->Identifiers in Python are names used to identify variables, functions, classes, modules, and other objects. They serve as labels to refer to these objects within a program. Here are the rules for naming identifiers in Python:

Valid Characters: Identifiers can consist of letters (both uppercase and lowercase), digits, and underscores (_). They must start with a letter or an underscore, but cannot start with a digit. Python is case-sensitive, so myVar and myvar are considered different identifiers.

Reserved Words: Python has reserved words (keywords) that cannot be used as identifiers because they have special meanings in the language. Examples of reserved words include if, else, for, while, def, class, and many more. You cannot use reserved words as identifiers.

Length: Identifiers can be of any length, but it is recommended to keep them meaningful and reasonably concise for readability.

PEP 8 Guidelines: Python follows PEP 8 style guidelines, which recommend using lowercase letters and underscores for variable and function names (snake_case). Class names should start with an uppercase letter (CamelCase).

Avoid Confusing Characters: It is best to avoid using characters that are easily confused, such as lowercase "l" (L), uppercase "O" (o), and uppercase "I" (i), especially in combination with numbers.

```python
def sum_of_digits(num):
    while num >= 10:
        sum_digits = 0
        while num > 0:
            sum_digits += num % 10
            num //= 10
        num = sum_digits
    return num

# Example usage
number = int(input("Enter a number: "))
result = sum_of_digits(number)
print("The sum of digits reduced to a single digit:", result)
```

```python
def calculate_lcm(num1, num2):
    greater = max(num1, num2)
    while True:
        if greater % num1 == 0 and greater % num2 == 0:
            lcm = greater
            break
        greater += 1
    return lcm

def calculate_hcf(num1, num2):
    while num2 != 0:
        num1, num2 = num2, num1 % num2
    return num1

# Example usage
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

lcm = calculate_lcm(number1, number2)
hcf = calculate_hcf(number1, number2)

print("The LCM of", number1, "and", number2, "is", lcm)
print("The HCF of", number1, "and", number2, "is", hcf)
```

```python
def print_cube_divisible_by_three(n):
    for num in range(1, n + 1):
        if num % 3 != 0:
            continue
        cube = num ** 3
        print(f"The cube of {num} is: {cube}")

# Example usage
number = int(input("Enter a number: "))
print("Using a continue statement:")
print_cube_divisible_by_three(number)
```

```python
def validate_roll_number(roll):
    if roll >= 1000 and roll <= 9999:
        return True
    return False

# Function to calculate total marks
def calculate_total_marks(marks):
    total = 0
    for subject in marks:
        if subject >= 40:
            total += subject
        else:
            return 0
    return total

# Input data for 10 students
students_data = []
for i in range(10):
    while True:
        roll_number = int(input(f"Enter roll number for student {i+1}: "))
        if validate_roll_number(roll_number):
            break
        else:
            print("Invalid roll number! Roll number should be in the range 1000-9999.")

    marks = []
    for j in range(3):
        marks.append(int(input(f"Enter marks for subject {j+1} for student {i+1}: ")))

    students_data.append((roll_number, marks))

# Calculating total marks and other required data
total_greater_than_200 = 0
highest_total = 0
highest_total_roll_numbers = []

for roll_number, marks in students_data:
    total_marks = calculate_total_marks(marks)

    if total_marks > 200:
        total_greater_than_200 += 1

    if total_marks > highest_total:
        highest_total = total_marks
        highest_total_roll_numbers = [roll_number]
    elif total_marks == highest_total:
        highest_total_roll_numbers.append(roll_number)

# Printing the results
print(f"\nNumber of students with total marks greater than 200: {total_greater_than_200}")
print("Roll numbers of students with the highest total marks:")
for roll_number in highest_total_roll_numbers:
    print(roll_number)
```

```python
def count_digits(number):
    count = 0
    if number == 0:
        return 1
    while number > 0:
        count += 1
        number //= 10
    return count

# Example usage
number = int(input("Enter a number: "))
digit_count = count_digits(number)
print("Number of digits:", digit_count)
```

```python
def reverse_number(number):
    reverse = 0
    while number > 0:
        digit = number % 10
        reverse = (reverse * 10) + digit
        number //= 10
    return reverse

# Example usage
number = int(input("Enter a number: "))
reverse = reverse_number(number)
double_reverse = reverse * 2
print("Reverse of the number:", reverse)
print("Double of the reverse number:", double_reverse)
```
